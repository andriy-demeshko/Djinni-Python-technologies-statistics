```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
import os
```

```python
df = pd.read_csv("./technologies.csv")
df_columns = ["title", "experience", "technologies", "date", "views", "applications"]
df = df[df_columns]
df.head()
```

```python
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])
df["experience"] = df["experience"].astype(np.int8)
df = df[df["technologies"].notna()]
df.info()
```

```python
technologies_0_year = df[df["experience"] == 0]["technologies"].str.get_dummies(",")
if not technologies_0_year.empty:
    top_technologies_0_year = technologies_0_year.sum().nlargest(10)
top_technologies_0_year
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(top_technologies_0_year.index, top_technologies_0_year.values)
plt.title("Most demanded technologies for 0 years experience")
plt.xlabel("Technologies")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
fig.savefig(os.path.join(
    "djinni_analysis/diagrams",
    f"{datetime.today().strftime('%Y-%m-%d_%H-%M-%S')}_0_years.png"),
    bbox_inches="tight",
)
```
